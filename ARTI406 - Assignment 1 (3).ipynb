# Assignment 1: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')


df.loc[df['price'].str.contains('Free', case=False, na=False), 'price'] = '0'


df['price'] = df['price'].str.findall(r'\$?(\d+\.\d+)').str[-1]


df['price'] = pd.to_numeric(df['price'], errors='coerce')

print(df[['name', 'price']].head(10))
print(df.dtypes)

df.dtypes
```

```python
# Statistical summary
df.describe(include='all')
```

### Descriptive summary Statistics

### Univariate Analysis

```python
df_clean = df[(df['price'] > 0) & (df['price'] < 100)]
plt.figure(figsize=(8,5))
sns.histplot(df_clean['price'], bins=20)
plt.title("Distribution of Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

The histogram shows that most games are concentrated in the low price range (between 0 and 10). The frequency decreases as the price increases, with very few games above 50. This indicates a right-skewed distribution.

### Distribution of Release Years

```python

df['release_year'] = df['date'].dt.year
plt.figure(figsize=(8,5))
sns.histplot(df['release_year'].dropna(), bins=15)
plt.title("Distribution of Release Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

The histogram shows the distribution of game release years. Most games were released between 2015 and 2020, while very few games were released before 2010. This indicates that game production has increased significantly in recent years.

```python
df['name_length'] = df['name'].str.len()

plt.figure(figsize=(8,5))
sns.histplot(df['name_length'], bins=20)
plt.title("Distribution of Game Name Length")
plt.xlabel("Name Length")
plt.ylabel("Count")
plt.show()
```

The histogram shows that most game titles have a moderate length, typically between 10 and 30 characters. Longer game names are less common, resulting in a right-skewed distribution.

```python
df['developer'] = df['developer'].replace('-', np.nan)
df['developer'] = df['developer'].dropna()

top_dev = df['developer'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_dev.values, y=top_dev.index)
plt.title("Top 10 Developers by Number of Games")
plt.xlabel("Number of Games")
plt.ylabel("Developer")
plt.show()
```

The chart shows the top developers by number of games. A few developers have many games, while most have fewer games.

The histogram shows that most game titles have a moderate length, typically between 10 and 30 characters. Longer game names are less common, resulting in a right-skewed distribution.

```python
df['price_level'] = pd.cut(df['price'],
                          bins=[0, 10, 30, 100],
                          labels=['Cheap', 'Medium', 'Expensive'])

price_counts = df['price_level'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=price_counts.index, y=price_counts.values)
plt.title("Game Price Categories")
plt.xlabel("Price Level")
plt.ylabel("Count")
plt.show()
```

The chart shows that most games fall into the low price category, while fewer games are medium or expensive.

## Bivariate Analysis

### Average price by Release Year

```python
df_clean = df[(df['price'] > 0) & (df['price'] < 100)]
avg_price_by_year = df_clean.groupby('release_year')['price'].mean().sort_index()
avg_price_by_year = avg_price_by_year[avg_price_by_year.index >= 2000]
plt.figure(figsize=(10,5))
avg_price_by_year.plot(kind='bar')
plt.title("Average Price by Release Year")
plt.xlabel("Release Year")
plt.ylabel("Average Price")
plt.show()
```

The chart shows how the average game price changes over the years. Prices remain relatively stable, with slight increases in some recent years.

### Top Catgories by Number of Games

```python
import re

def split_categories(text):
    text = str(text).strip()
    if text == '' or text.lower() == 'nan':
        return []

    text = re.sub(r'(?<=[a-z])(?=[A-Z])', ';', text)
    
    parts = [x.strip() for x in text.split(';') if x.strip() != '']
    return parts

df['categories_list'] = df['categories'].apply(split_categories)
df_exploded = df.explode('categories_list')
top_cat = df_exploded['categories_list'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_cat.values, y=top_cat.index)
plt.title("Top Categories by Number of Games")
plt.xlabel("Number of Games")
plt.ylabel("Category")

plt.show()
```

The chart shows the most common categories in the dataset after separating the combined category text into individual categories. Single-player appears most frequently, followed by features such as Steam Achievements and Profile Features Limited.

### Top developer by price

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df_clean = df.dropna(subset=['developer', 'price'])
dev_revenue = df_clean.groupby('developer')['price'].sum().sort_values(ascending=False)
dev_revenue.head(10)
```

```python

df['price'] = pd.to_numeric(df['price'], errors='coerce')
df_clean = df.dropna(subset=['developer', 'price'])
df_clean = df_clean[(df_clean['price'] > 0) & (df_clean['price'] < 100)]
top_dev = df_clean.groupby('developer')['price'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_dev.values, y=top_dev.index)

plt.title("Top 10 Developers by Total Price")
plt.xlabel("Total Price")
plt.ylabel("Developer")
plt.show()
```

The chart shows the top 10 developers based on total game prices. Some developers clearly have higher total values compared to others.

### Correlation Matrix

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
df_corr = df[['price', 'release_year', 'name_length']].dropna()
print("Rows left:", len(df_corr))
plt.figure(figsize=(6,4))
sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

- Correlation close to 1 = strong positive relationship
- Close to 0 = weak relationship

## Time-Based Analysis
### Game Release by Month

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['month'] = df['date'].dt.month
df_clean = df.dropna(subset=['month'])
monthly_games = df_clean.groupby('month')['name'].count().sort_index()
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.figure(figsize=(10,5))
plt.plot(monthly_games.index, monthly_games.values, marker='o')

plt.xticks(range(1, 13), month_labels)
plt.title("Game Releases by Month")
plt.xlabel("Month")
plt.ylabel("Number of Games")
plt.show()
```
